## importing required  libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## coverting into dataframe

```python
d1=pd.read_table("student-mat.csv",sep=";")
d2=pd.read_table("student-por.csv",sep=";")

frames = [d1,d2] 

d3 = pd.concat(frames)
```

## exploring the range of values for numeric values

```python
d3.describe()
```

```python
categorical_columns = d3.select_dtypes(include=['int','float']).columns
for column in categorical_columns:
    distinct_values = d3[column].unique()
    print(f"{column} : {distinct_values}")
```

## exploring the range of values for categorical values

```python
categorical_columns = d3.select_dtypes(include=['object']).columns
for column in categorical_columns:
    distinct_values = d3[column].unique()
    print(f"{column} : {distinct_values}")
```

## distribution of each columns

```python
for column in d3.columns:
    plt.figure()
    if pd.api.types.is_numeric_dtype(d3[column]):
        # Numeric column
        d3[column].hist()
        plt.title(f"Distribution of values in column '{column}'")
        plt.xlabel("Values")
        plt.ylabel("Frequency")
    else:
        # Categorical column
        value_counts = d3[column].value_counts()
        value_counts.plot(kind='bar')
        plt.title(f"Distribution of values in column '{column}'")
        plt.xlabel("Distinct values")
        plt.ylabel("Frequency")
        plt.xticks(rotation=0)
                  
    plt.show()
```
